# Phase 1

```python

# Import Library
import pandas as pd
import numpy as np
```

```python
# For Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
td = pd.read_csv("train_data.csv")
td.head()
```

```python
#Getting overview about the data 
td.info()
```

```python
#Statistical Analysis on Numerical Data
td.describe().T
```

```python
# Missing Values

print (td.isnull().sum())

# Visualize the missing data
plt.figure(figsize = (10,3))
sns.heatmap(td.isnull(), cbar=True, cmap="magma")
```

```python
#Statistical Analysis on categorical Data
td.describe(include=['object','bool'])
```

```python
#dropping off some redundant features
td.drop(['Over18','StandardHours','EmployeeCount','EmployeeNumber'],axis=1,inplace=True)
```

```python
#creating a copy of the dataset for part B
td2 = td.copy()
```

```python
td2.head()
```

```python
#encode the categorical features to numerical ones
td2 = pd.get_dummies(td2,drop_first=True)
```

```python
#segment dataset into data and target label
target = td2.pop("Attrition")
```

```python
# scale dataset features

from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
scaler = MinMaxScaler()
td_scaled = pd.DataFrame(scaler.fit_transform(td2), columns = td2.columns)
td_scaled
```

```python
#identifying key featurs from the data set
model = RandomForestClassifier()
```

# Well you can use a random classifier model to identify the importance features


```python
# fit the model
model.fit(td_scaled, target)
importances = model.feature_importances_
sort_imp = np.argsort(importances)
names = list(td2.columns)
```

```python
#plotting a feature importance chart
plt.figure(figsize=(10,7))
plt.barh(range(len(sort_imp)),importances[sort_imp], color="red")
plt.yticks(range(len(sort_imp)), [names[x] for x in sort_imp])
plt.title("Feature Importance")
plt.xlabel("relative importance measure")
```

```python
#Now we have cleaned the data(Pre-processed), removed redundant features for this computation and visualized if we had missing values. 
```

# Phase 2

```python
# split the DataFrame into train and test datasets

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_val = train_test_split(td_scaled, target, train_size=0.8, random_state=1)
x_train # This is the data we will be training from
```

```python
x_test # This is the data we will be predicting from. 
```

```python
y_train
```

```python
# Import model classifier 

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC 
```

```python
# Instantiate model
log_reg = LogisticRegression()
```

```python
# Fit a linear regression model on the training test which is the 80% of data you splitted
#fit model(training model)

log_reg.fit(x_train,y_train)
```

```python
# now let's predict with the x_test 20% dataset which is the testing data

log_pred = log_reg.predict(x_test)
log_pred
```

# Phase 3

```python
# import evaluation metrics

from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Evaluate the model using the test data

print(classification_report(y_val, log_pred))
```

```python
# Confusion matrix or Clarification report
# This plots the graph that tells us the number of prediction in TP, FP or TN, FN

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
cm = confusion_matrix(y_val,log_pred)
labels = np.array(
    [
        ["{0:0.0f}".format(item) + "\n{0:.2%}".format(item / cm.flatten().sum())]
        for item in cm.flatten()
    ]
).reshape(2,2)

plt.figure(figsize=(6,4))
sns.heatmap(cm,annot=labels,fmt="");
plt.ylabel("True labels");
plt.xlabel("Predicted labels");
```

# Phase 4

```python
# Hyper Parameters Optimization
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, r2_score
```

```python
# Use a Gradient Boosting algorithm

alg = GradientBoostingClassifier()
```

```python
# Try these hyperparameter values

params = {
    'learning_rate': [0.1,0.5,1],
    'n_estimators' : [50,100,150]
}
```

```python
# Find the best hyperparameter combination to optimize the R2 metric

score = make_scorer(r2_score)
grid = GridSearchCV(alg,params,scoring=score, cv=3)
grid.fit(x_train,y_train)
print("Best parameter combination:", grid.best_params_, "\n")
```

```python
# measuring performance on test set

print ("Applying best model on test data:")
best_mod = grid.best_estimator_
pred = best_mod.predict(x_test)
```

```python
# Now let's evaluate the model using the test data

print(classification_report(y_val, pred))
```

# Phase 5

```python
# Using the Retraining Model

# Load and prepare the data again

td = pd.read_csv("train_data.csv")
td.drop(['Over18','StandardHours','EmployeeCount','EmployeeNumber'],axis=1,inplace=True)
td2 = td.copy()
td2 = pd.get_dummies(td2,drop_first=True)
target = td2.pop("Attrition")
x_train, x_test, y_train, y_val = train_test_split(td2, target, train_size=0.8, random_state=1)


```

```python
# Define the model and hyperparameter grid

alg = GradientBoostingClassifier()
params = {
    'learning_rate': [0.1,0.5,1],
    'n_estimators' : [50,100,150]
}
```

```python
# Create a scoring function

score = make_scorer(r2_score)

```

```python
# Perform grid search

grid = GridSearchCV(alg,params,scoring=score, cv=3)
grid.fit(x_train,y_train)
```

```python
# Identify the best model
best_mod = grid.best_estimator_
```

```python
# Re-train the model on the entire training data

best_mod.fit(td2, target)
```

```python
# Evaluate the re-trained model on the test data

pred = best_mod.predict(x_test)
print(classification_report(y_val, pred))
```

```python
# Now letâ€™s Plot the New Confusion Metrics 

#Library Importation

from sklearn.metrics import confusion_matrix, roc_curve, auc
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have your predicted labels (`pred`) and true labels (`y_val`) as used in this computation I created. 


# Confusion Matrix

cm = confusion_matrix(y_val,pred)

labels = np.array(
    [
        ["{0:0.0f}".format(item) + "\n{0:.2%}".format(item / cm.sum())]
        for item in cm.flatten()
    ]
).reshape(2,2)


# Plot confusion matrix

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=labels, fmt="", cbar=False, cmap='Blues');
plt.ylabel("True Labels")
plt.xlabel("Predicted Labels")
plt.title("Confusion Matrix")
plt.show()
```

```python
# Now letâ€™s Plot an ROC Curve

fpr, tpr, _ = roc_curve(y_val, pred)
roc_auc = auc(fpr, tpr)


# Plot ROC curve


plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--', label='No Skill')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
```
